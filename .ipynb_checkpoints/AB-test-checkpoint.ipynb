```python
def random_users_split(df):
    user_ids = df.index.unique()
    users_A = np.random.choice(user_ids, size=len(user_ids) // 2, replace=False)
    users_B = np.setdiff1d(user_ids, users_A)
    return users_A, users_B
```

```python
def plot_tms(df_train_all, df_valid_all, users_A, users_B, name):
    df_train_avg_A = df_train_all.loc[users_A].groupby(['day']).mean().metric
    df_train_avg_B = df_train_all.loc[users_B].groupby(['day']).mean().metric
    df_valid_avg_A = df_valid_all.loc[users_A].groupby(['day']).mean().metric
    df_valid_avg_B = df_valid_all.loc[users_B].groupby(['day']).mean().metric
    plt.figure(figsize=(25,20))
    plt.plot(df_train_avg_A.index, df_train_avg_A, label='train A')
    plt.plot(df_train_avg_B.index, df_train_avg_B, label='train B')
    plt.plot(df_valid_avg_A.index, df_valid_avg_A, label='valid A')
    plt.plot(df_valid_avg_B.index, df_valid_avg_B, label='valid B')
    plt.ylabel('Avg check')
    plt.xlabel('day')
    plt.legend()
    plt.savefig(name)
```

```python
users_A, users_B = random_users_split(df)
```

```python
plot_tms(df_train_all, df_valid_all, users_A, users_B, 'AVG groups')
```

```python
def cuped(df_all_A_train_reg, df_all_B_train_reg, df_all_A_valid_reg, df_all_B_valid_reg):

    df_all_A = df_all_A_train_reg.merge(df_all_A_valid_reg, on = ['user_id'])
    df_all_B = df_all_B_train_reg.merge(df_all_B_valid_reg, on = ['user_id'])
    df_all_A['group'] = 'A'
    df_all_B['group'] = 'B'
    
    df_all_cuped = pd.concat([df_all_A, df_all_B], axis = 0)
    df_all_cuped = shuffle(df_all_cuped)
    
    mean_cov = df_all_cuped['metric_x'].mean()
    variance = df_all_cuped['metric_x'].var()
    cov = df_all_cuped[['metric_x','metric_y']].cov().values[0,1]
    theta = cov / variance
    df_all_cuped['cuped'] = df_all_cuped.apply(lambda x: x['metric_y'] - (x['metric_x'] - mean_cov) * theta, axis = 1)
    
    df_all_A = df_all_cuped[df_all_cuped.group == 'A']
    df_all_B = df_all_cuped[df_all_cuped.group == 'B']
    return df_all_A, df_all_B
```

```python
def mw_k2s(iters, df_train_all, df_valid_all, k_mw, k_k2s, k_mean, effect, df, k_cuped):
    p_values = []
    similar_groups = []
    for i in range(iters):
        users_A, users_B = random_users_split(df)
        df_t_A, df_t_B = df_train_all.loc[users_A], df_train_all.loc[users_B]
        t_A, t_B = df_t_A.metric, df_t_B.metric
        p_k2s = scipy.stats.ks_2samp(t_A, t_B).pvalue
        p_mw = scipy.stats.mannwhitneyu(t_A, t_B, alternative='two-sided').pvalue
        mean = abs(t_A.mean() - t_B.mean())

        if p_mw > k_mw and p_k2s > k_k2s and mean < k_mean:
            similar_groups.append([users_A, users_B])
            df_v_A, df_v_B = df_valid_all.loc[users_A], df_valid_all.loc[users_B]
            v_A = df_v_A.metric
            
            if effect != 1:
                df_v_B.metric = df_v_B.metric * effect
            v_B = df_v_B.metric
            if k_cuped == 1:
                df_all_A, df_all_B = cuped(df_t_A, df_t_B, df_v_A, df_v_B)
                p_ans = scipy.stats.mannwhitneyu(df_all_A['cuped'].tolist(), df_all_B['cuped'].tolist()).pvalue
            else:
                p_ans = scipy.stats.mannwhitneyu(v_A, v_B).pvalue
            p_values.append(p_ans)
    return p_values, similar_groups
```

```python
iters = 5000

p_values, similar_groups = mw_k2s(iters,df_train_all, df_valid_all, 0.2, 0.2, 1000, 1, df, 0)
```

```python
p_values = np.array(p_values)
p_values[p_values < 0.05].shape[0] / len(similar_groups)
```

## Мощность для тестов длительностью в N число дней.

```python
iters = 500

effect = 1.02

powers = []


for n in range(1, 11):
    df_t_n = df_train_all[df_train_all['day'] < n]
    df_v_n = df_valid_all[df_valid_all['day'] < 10 + n]

    p_values, similar_groups = mw_k2s(iters,df_t_n, df_v_n, 0.2, 0.2, 1000, effect, df_t_n, 0)
    
    p_values = np.array(p_values)
    power = p_values[p_values < 0.05].shape[0] / len(similar_groups)
    powers.append(power)
```

```python
pd.DataFrame({'Мощность': powers})
```

```python
iters = 1000

effect = 1.02

powers = []
for n in range(1, 11):
    df_t_n = df_train_all[df_train_all['day'] < n]
    df_v_n = df_valid_all[df_valid_all['day'] < 10 + n]

    p_values, similar_groups = mw_k2s(iters,df_t_n, df_v_n, 0.2, 0.2, 1000, effect, df_t_n, 1)
    
    p_values = np.array(p_values)
    power = p_values[p_values < 0.05].shape[0] / len(similar_groups)
    powers.append(power)
```

```python
pd.DataFrame({'Мощность': powers})
```
